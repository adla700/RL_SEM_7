```python
import numpy as np
import matplotlib.pyplot as plt

class MultiArmedBandit:
    def __init__(self, num_arms, epsilon=0.1):
        self.num_arms = num_arms
        self.epsilon = epsilon
        self.q_values = np.zeros(num_arms)
        self.num_pulls = np.zeros(num_arms)

    def choose_action(self):
        if np.random.random() < self.epsilon:
            # Exploration: Choose a random arm
            return np.random.randint(self.num_arms)
        else:
            # Exploitation: Choose the arm with the highest estimated Q-value
            return np.argmax(self.q_values)

    def update(self, action, reward):
        self.num_pulls[action] += 1
        self.q_values[action] += (1.0 / self.num_pulls[action]) * (reward - self.q_values[action])

def run_simulation(num_arms, num_steps, epsilon):
    bandit = MultiArmedBandit(num_arms, epsilon)
    rewards = []

    for step in range(num_steps):
        action = bandit.choose_action()
        reward = np.random.binomial(1, 0.5)  # Replace with a reward distribution specific to recommendation system
        bandit.update(action, reward)
        rewards.append(reward)

    return rewards

# Example usage
num_arms = 10
num_steps = 1000
epsilon = 0.45

rewards = run_simulation(num_arms, num_steps, epsilon)
plt.plot(rewards)
plt.xlabel("Step")
plt.ylabel("Reward")
plt.title("Multi-Armed Bandit Simulation")
plt.show()
```
